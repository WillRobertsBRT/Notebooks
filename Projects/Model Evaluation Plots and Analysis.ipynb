```python
import scipy.stats as sp
from matplotlib import pyplot as plt, rcParams
import seaborn as sns
import pandas as pd
import numpy as np

sns.set_style('darkgrid')
sccc = pd.read_csv('soil_color_cloud_cover.csv')

dark = sccc[sccc['soil_color']=='DARK']
light = sccc[sccc['soil_color']=='LIGHT']
cloudy = sccc[sccc['cloud_cover']=='OVERCAST']
clear = sccc[sccc['cloud_cover']=='CLEAR']
rcParams['font.size']=16
plt.figure(figsize=(10,10))

s, p =sp.ttest_ind(dark['GridMetric_instance_recall'], light['GridMetric_instance_recall'])
print(f'dark/light p-value is {p:.3f}')

s, p =sp.ttest_ind(cloudy['GridMetric_instance_recall'], clear['GridMetric_instance_recall'])
print(f'clear/overcast p-value is {p:.3f}')

plt.scatter(dark['GridMetric_instance_precision'], dark['GridMetric_instance_recall'],marker  ='o', s=80, color='blue')
plt.scatter(light['GridMetric_instance_precision'], light['GridMetric_instance_recall'],marker  ='o', s=80, color='darkorange')
plt.xlabel('Grid Metric Precision')
plt.ylabel('Grid Metric Recall')
plt.xlim(0,1.1)
plt.ylim(0,1.1)
plt.title('soybeans_0528_1 Performance on Light vs Dark Soil Color Fields')
plt.legend(['Dark Soil', 'Light Soil'])
plt.show()

plt.figure(figsize=(10,10))
plt.scatter(cloudy['GridMetric_instance_precision'], cloudy['GridMetric_instance_recall'],marker  ='o', s=100, color='blue')
plt.scatter(clear['GridMetric_instance_precision'], clear['GridMetric_instance_recall'],marker  ='o', s=100, color='darkorange')
plt.xlabel('Precision')
plt.ylabel('Recall')
plt.xlim(0,1.1)
plt.ylim(0,1.1)
plt.legend(['Overcast', 'Clear'])
plt.title('soybeans_0528_1 Performance on Cloudy vs Overcast Fields')
plt.show()
```
